# K-Means

```python
%matplotlib inline 
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 
from sklearn.cluster import KMeans
```

```python
dataset.keys()
```

```python
dataset.info()
```

```python
dataset.head()
```

```python
X = np.asarray(dataset)
```

```python
plt.scatter(X[:,1],X[:,2], label = 'True Position')
plt.xlabel("Sebelum Lebaran")
plt.ylabel("Setelah Lebaran")
plt.title("Perubahan Harga KOMODITAS")
```

```python
Xn=X[:,1:3]
Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(Xn)
    Sum_of_squared_distances.append(km.inertia_)

plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(Xn)
```

```python
print(kmeans.cluster_centers_)
```

```python
print(kmeans.labels_)
```

```python
plt.scatter(X[:,1],X[:,2], c=kmeans.labels_, cmap='rainbow')
plt.xlabel("Sebelum Lebaran")
plt.ylabel("Setelah Lebaran")
plt.title("Perubahan Harga")
plt.show()
```

```python
plt.scatter(X[:,1],X[:,2], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='black')
plt.xlabel("Sebelum Lebaran")
plt.ylabel("Setelah Lebaran")
plt.title("Perubahan Harga")
plt.show()
```

```python
# Memasukkan hasil label ke dalam suatu 
data_clustered = pd.DataFrame(Xn)
data_clustered['cluster'] = pd.DataFrame(kmeans.labels_)
data_clustered = data_clustered.sort_values('cluster')
data_clustered = np.asarray(data_clustered)
```

```python
# Membuat similarity Matrix 
from scipy.spatial.distance import cdist
dist = cdist(data_clustered[:,0:2],data_clustered[:,0:2], metric='euclidean')
plt.imshow(dist)
plt.show()
```

```python
from sklearn import datasets 
from sklearn.cluster import KMeans 
from sklearn.metrics import davies_bouldin_score 
from sklearn.datasets.samples_generator import make_blobs 

# we store the cluster labels 
  
print(davies_bouldin_score(Xn, kmeans.labels_)) 
```

# Hierarchical

```python
%matplotlib inline 
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
dendrogram = sch.dendrogram(sch.linkage(Xn, method='ward'))
```

```python
model = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
model.fit(Xn)
labels = model.labels_
labels
```

```python
plt.scatter(Xn[labels==0, 0], Xn[labels==0, 1], s=50, marker='o', color='red')
plt.scatter(Xn[labels==1, 0], Xn[labels==1, 1], s=50, marker='o', color='blue')
plt.scatter(Xn[labels==2, 0], Xn[labels==2, 1], s=50, marker='o', color='green')
plt.show()
```

```python
data_clustered2 = pd.DataFrame(Xn)
data_clustered2['cluster'] = pd.DataFrame(model.labels_)
data_clustered2 = data_clustered2.sort_values('cluster')
data_clustered2 = np.asarray(data_clustered2)
```

```python
from scipy.spatial.distance import cdist
dist = cdist(data_clustered2[:,0:2],data_clustered2[:,0:2], metric='euclidean')
plt.imshow(dist)
plt.show()
```
